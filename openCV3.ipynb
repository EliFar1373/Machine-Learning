```python
#norm 2
import numpy as np
def normalized_image(img):
   
    

    norm_img=cv2.normalize(img,None,100,160,cv2.NORM_MINMAX)
    return norm_img
norm_im=normalized_image(img)
plt.imshow(norm_im)
```

```python
#Image Filtering using convolutional kernels
#identity kernel
kernel=np.array([[0,0,0],[0,1,0],[0,0,0]])
img_identity=cv2.filter2D(img,-1,kernel)
plt.imshow(img_identity)
```

```python
#blur kernel
#Box Blur (Averaging Filter):
#sum of elemnt shouold be one s
kernel=np.ones((5,5),np.float32)/25
blur_img=cv2.filter2D(img,-1,kernel)
plt.imshow(blur_img)
```

```python
'''
Median blur is a type of non-linear filtering. 
It replaces each pixel value with the median value of its neighborhood.
src is the image file, ksize is the kernel size - the size of neighborhood window. It must be an odd integer.
'''

img_blur_median=cv2.medianBlur(img,5)
plt.imshow(img_blur_median)
```

```python
#Image Sharpening with a kernel
kernel=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
img_sharpening=cv2.filter2D(img,-1,kernel)
plt.imshow(img_sharpening)
```

```python
'''
Image thresholding is a common image processing technique used to separate objects 
or regions of interest from the background by converting a grayscale image into a binary image. 
'''
img=cv2.imread(r'C:\Users\elham\Desktop\image\train.jpg')
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,img_white_black=cv2.threshold(img_gray,150,200,cv2.THRESH_BINARY)
plt.imshow(img_white_black,cmap='gray')
```

```python
#Edge detection using Sobel Operator and canny

# Convert to graycsale
img1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to the grayscale image using cv2.GaussianBlur 
# with a kernel size of (3,3) to smooth the image and reduce noise.
blur_img=cv2.GaussianBlur(img1,(3,3),0,0)

# Sobel Edge Detection
 # Horizontal Edges
sobelx=cv2.Sobel(blur_img,ddepth=cv2.CV_64F,dx=1,dy=0,ksize=5)
# Vertical Edges
sobely=cv2.Sobel(src=blur_img,ddepth=cv2.CV_64F,dx=0,dy=1,ksize=5)

sobelxy = cv2.Sobel(src=blur_img,ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Horizontal and vertical edges


#Canny Edge detection
canny_img=cv2.Canny(blur_img,threshold1=100,threshold2=200)

# Assuming sobelxy, sobelx, and sobely are defined
fig, ax = plt.subplots(2, 3, figsize=(15, 5))
ax[0,0].imshow(sobelxy,cmap='gray')
plt.axis('off')
ax[0,1].imshow(sobelx,cmap='gray')
plt.axis('off')
ax[0,2].imshow(sobely,cmap='gray')
plt.axis('off')
ax[1,0].imshow(canny_img,cmap='gray')
```
