```python
x=df.drop('income',axis=1)
y=df['income']
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
x_train.shape,x_test.shape
```

```python
for var in x_train.columns:
    x_train[var].fillna(x_train[var].mode()[0],inplace=True)
    
x_train.isnull().sum()
```

```python

    
for var in x_test.columns:
    x_test[var].fillna(x_test[var].mode()[0],inplace=True)
    
x_test.isnull().sum()
```

```python
x_train_categorical=[var for var in x_train.columns if x_train[var].dtype=='object']
x_train[x_train_categorical].head(2)
```

```python
x_train.info()
```

```python
import pandas as pd
x_train_df=pd.get_dummies(x_train,columns=x_train_categorical)
x_test_df=pd.get_dummies(x_test,columns=x_train_categorical)
x_test_df.head(2)
```

```python

x_train_df.shape
```

```python
# Convert the resulting True/False values to 1/0
x_train_df = x_train_df.astype(int)
x_test_df = x_test_df.astype(int)
```

```python
x_test_df.columns.equals(x_train_df.columns)
```

```python
# Ensure the columns are the same by reindexing x_test_df to match x_train_df's columns
x_test_df = x_test_df.reindex(columns=x_train_df.columns, fill_value=0)
```

```python
x_test_df.head(2)
```

```python
#11. Feature Scaling
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
col=x_train_df.columns
x_train_df2=scaler.fit_transform(x_train_df)
x_test_df2=scaler.transform(x_test_df)

x_train_df2=pd.DataFrame(x_train_df2,columns=col)
x_test_df2=pd.DataFrame(x_test_df2,columns=col)

x_train_df2.head(2)
```
