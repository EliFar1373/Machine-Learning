```python
import matplotlib.pyplot as plt
import cv2

img=cv2.imread(r'C:\Users\elham\OneDrive\Desktop\image\1.jpg')

b,g,r=cv2.split(img)
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img3=cv2.GaussianBlur(img,(11,11),0)
img4=cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
img5=cv2.medianBlur(img,7)
img6=cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)
img7=cv2.bilateralFilter(img,7,90,79)
img8=cv2.cvtColor(img7,cv2.COLOR_BGR2RGB)

img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret, thresh=cv2.threshold(img2,127,255,cv2.THRESH_BINARY)


plt.imshow(thresh,cmap='gray')








#plt.imshow(b,cmap='Blues')
```
